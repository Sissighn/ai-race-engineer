```python
from src.data.load_data import load_session, load_telemetry
from src.data.feature_engineering import segment_corners, build_features
from src.data.compare import sync_telemetry   

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

```python
session = load_session(2023, "Silverstone", "Q")

tel_VER = load_telemetry(session, "VER")
tel_PER = load_telemetry(session, "PER")

print("VER shape:", tel_VER.shape)
print("PER shape:", tel_PER.shape)
```

```python
df = sync_telemetry(tel_VER, tel_PER)

bool_cols = ["Brake_1", "Brake_2", "DRS_1", "DRS_2"]

for col in bool_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)
        
df["Speed_Delta"] = df["Speed_1"] - df["Speed_2"]
df["Throttle_Delta"] = df["Throttle_1"] - df["Throttle_2"]
df["Brake_Delta"] = df["Brake_1"] - df["Brake_2"]

df.head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(df["Distance"], df["Speed_1"], label="VER")
plt.plot(df["Distance"], df["Speed_2"], label="PER")
plt.title("Speed Comparison VER vs PER")
plt.xlabel("Distance (m)")
plt.ylabel("Speed (km/h)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.plot(df["Distance"], df["Speed_Delta"], color="purple")
plt.axhline(0, color="black", linewidth=1)
plt.title("Delta Speed (VER - PER)")
plt.xlabel("Distance (m)")
plt.ylabel("Delta (km/h)")
plt.grid(True)
plt.show()
```
